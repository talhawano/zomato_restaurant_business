### Removing_missing value ###

**here we are removing the '/5', nan values to "/" and object column to float**

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate']= df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

***above i have used fill na function to fill the missing value with mean of the rate column***

### Dropping null values and renaming columns ###
for another columns which are less usefull

```python
df.info()
```

```python
df.dropna(inplace=True)
df.columns
```

```python
df.rename(columns = {'approx_cost(for two people)':'cost2plates', 'listed_in(type)':'type'}, inplace=True)
df.head()
```

```python
df['listed_in(city)'].unique()
```

*** you can merge or keep only one column ,although it is same so we keep location as city has less details ***

```python
df = df.drop(['listed_in(city)'], axis=1)
```

```python
df['cost2plates'].unique()
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
         return float(value)

df['cost2plates'] = df['cost2plates'].apply(handlecomma)
df['cost2plates'].unique()
```

```python
df.dtypes
```

```python
df['rest_type'].value_counts() 
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_types(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(handle_rest_types)
df['rest_type'].value_counts()
```

### Cleaning column as required ###

```python
df.cuisines.value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending = False)
cuisines_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return'others'
    else:
        return value
    
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
